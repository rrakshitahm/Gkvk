```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a dummy DataFrame for demonstration purposes
# You should replace this with your actual data loading code (e.g., pd.read_csv('your_data.csv'))
data = {
    "Yield_kg": np.random.rand(10) * 100,
    "Soil_Moisture": np.random.rand(10) * 50,
    "Zone": np.random.randint(1, 4, 10)
}
df = pd.DataFrame(data)

plt.scatter(df["Yield_kg"],
 df["Soil_Moisture"],
 c=df["Zone"])
plt.xlabel("Yield (kg)")
plt.ylabel("Soil Moisture")
plt.title("Farm Field Clusters")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
X = df[["Yield_kg", "Soil_Moisture"]] # Inputs
y = df["Zone"] # Output (e.g., Zone)
model = LogisticRegression()
model.fit(X, y)
```

```python
# Predict probability
model.predict_proba([[3, 90]])
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X)
confusion_matrix(y, y_pred)
```
