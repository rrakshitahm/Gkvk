```python
import pandas as pd
# Plot-level farm data
data = {
 "Yield_kg": [3200, 3000, 3500, 2800, 3600, 2900],
 "Soil_Moisture": [30, 28, 35, 25, 36, 27]
}
df = pd.DataFrame(data)
df
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df["Zone"] = kmeans.fit_predict(df)
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Yield_kg"],
 df["Soil_Moisture"],
 c=df["Zone"])
plt.xlabel("Yield (kg)")
plt.ylabel("Soil Moisture")
plt.title("Farm Field Clusters")
plt.show()
```
