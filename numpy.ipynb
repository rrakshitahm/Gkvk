```python
import numpy as np


if 'print' in globals() and not callable(globals()['print']):
    del print

yields = np.array([
 [120, 125, 130, 135],
 [115, 120, 122, 128],
 [130, 135, 140, 142]
])


overall_avg = np.mean(yields)
print(f"Overall Average: {overall_avg}")

field_avgs = np.mean(yields, axis=1)
print(f"Field Averages: {field_avgs}")

year_avgs = np.mean(yields, axis=0)
print(f"Year Averages: {year_avgs}")
```
